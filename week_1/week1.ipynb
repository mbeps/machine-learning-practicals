```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.datasets import load_iris
```

# Classify Iris (Simple)

## Loading Data

```python
iris = load_iris()
X = iris.data [:, [1, 3]] # 1 and 3 are the features we will use here.
y = iris.target
```

## Classifier

```python
iris_tree = tree.DecisionTreeClassifier(criterion="entropy").fit(X, y)
```

## Prediction
Here we use the tree to predict the classification of each background point.

```python
Z = iris_tree.predict(np.c_[xx.ravel(), yy.ravel()])
```

## Plots

Now plot the decision surface that we just learnt by using the decision tree to classify every packground point.

```python
plot_step = 0.02

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1 # Get the boundaries of the data
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1 # Get the boundaries of the data
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

Z = Z.reshape(xx.shape) # Put the result into a color plot
cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)

# Also plot the original data on the same axes
plt.scatter(X[:, 0], X[:, 1], c=y.astype(float)) #, cmap='autumn')

# Label axes
plt.xlabel( iris.feature_names[1], fontsize=10 )
plt.ylabel( iris.feature_names[3], fontsize=10 )

plt.show()
```
